```python
def sympmotion(n):
    x = np.array([])
    v = np.array([])
    step = np.linspace(0,1,n)
    h = 10**(-2)
    for i in range(n):
        if i == 0:
            x = np.append(x, [0])
            v = np.append(v, [1])
        else:
            x = np.append(x, [x[i-1] + v[i-1]*h])
            v = np.append(v, [v[i-1] - x[i]*h])
            
    
    xplt = plt.plot(x,v)
    
    print(xplt)
```

```python
def twobody(xo,yo,vxo,vyo,t):
    x = np.array([])
    y = np.array([])
    vx = np.array([])
    vy = np.array([])
    ax = np.array([])
    ay = np.array([])
    
    #step = np.linspace(0,1,t)
    h = 10**(-4)
    for i in range(t):
        if i == 0:
            x = np.append(x, [xo])
            y = np.append(y, [yo])
            vx = np.append(vx, [vxo])
            vy = np.append(vy, [vyo])
            ax = np.append(ax, [0])
            ay = np.append(ay, [0])
        else:
            
            ax = np.append(ax, [(-1*G*Ms*x[i-1])/((x[i-1]**2 + y[i-1]**2)**(3/2))])
            ay = np.append(ay, [(-1*G*Ms*y[i-1])/((x[i-1]**2 + y[i-1]**2)**(3/2))])
            
            x = np.append(x, [x[i-1] + vx[i-1]*h])
            y = np.append(y, [y[i-1] + vy[i-1]*h])
            vx = np.append(vx, [vx[i-1] + ax[i]*h])
            vy = np.append(vy, [vy[i-1] + ay[i]*h])
            
            #if i == 25000:
             #   print(x[i],y[i],vx[i],vy[i])
            
    fig, ax = plt.subplots()
    ax.plot(x,y)
    ax.set_xlim(-2,2)
    ax.set_ylim(-2,2)
    plt.axis('equal')
    plt.scatter(0,0, c='orange')
    print(ax)
```
