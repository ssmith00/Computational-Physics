```python
G = 1.18312*(10**-4) #G in dimensions of AU, earth masses, and years.

Ms = 333060 #Mass of the sun
```

```python
import numpy as np
import matplotlib.pyplot as plt

class planet:
    def __init__(self, name, M, T, sma, R, xo,yo):
        self.name = name
        self.M = M
        self.T = T
        self.sma = sma
        self.R = R
        self.xo = xo
        self.yo = yo
```

```python

mer = planet("Mercury", 0.05528, 0.2411, 0.3871, (4.09*(10**-19)), 0.32622, 0.2084)
ven = planet("Venus", 0.8157, 0.6156, 0.7233, (1.01*(10**-18)), -0.0151,0.7231)
ear = planet("Earth",1,1,1,(1.07*(10**-18)),1,0)
mar = planet("Mars",0.1075,1.882,1.527,(5.68*(10**-19)),-0.2497,-1.507)
jup = planet("Jupiter",318.3,11.86,5.202,(1.17*(10**-17)),2.085,4.767)
sat = planet("Saturn",95.14,29.45,9.539,(9.75*(10**-18)),-1.232,9.459)
ura = planet("Uranus",14.54,84.02,19.19,(4.25*(10**-18)),-17.26,-8.383)
nep = planet("Neptune",17.09,164.8,30.06,(4.12*(10**-18)),-28.14,10.58)
```

```python
def orbital(t):
    
    merx = mer.sma*np.cos(2*np.pi*t/mer.T) + mer.xo
    venx = ven.sma*np.cos(2*np.pi*t/ven.T) + ven.xo
    earx = ear.sma*np.cos(2*np.pi*t/ear.T) + ear.xo
    marx = mar.sma*np.cos(2*np.pi*t/mar.T) + mar.xo
    jupx = jup.sma*np.cos(2*np.pi*t/jup.T) + jup.xo
    satx = sat.sma*np.cos(2*np.pi*t/sat.T) + sat.xo
    urax = ura.sma*np.cos(2*np.pi*t/ura.T) + ura.xo
    nepx = nep.sma*np.cos(2*np.pi*t/nep.T) + nep.xo
        
    #x = np.array([merx,venx,earx,marx,jupx,satx,urax,nepx])
     
    mery = mer.sma*np.sin(2*np.pi*t/mer.T) + mer.yo
    veny = ven.sma*np.sin(2*np.pi*t/ven.T) + ven.yo
    eary = ear.sma*np.sin(2*np.pi*t/ear.T) + ear.yo
    mary = mar.sma*np.sin(2*np.pi*t/mar.T) + mar.yo
    jupy = jup.sma*np.sin(2*np.pi*t/jup.T) + jup.yo
    saty = sat.sma*np.sin(2*np.pi*t/sat.T) + sat.yo
    uray = ura.sma*np.sin(2*np.pi*t/ura.T) + ura.yo
    nepy = nep.sma*np.sin(2*np.pi*t/nep.T) + nep.xo
    
    #y = np.array([mery,veny,eary,mary,jupy,saty,uray,nepy])
    
    fig, ax = plt.subplots(figsize=(30,30))
    
    ax.scatter(mery,merx, color='grey', s=10)
    ax.scatter(veny,venx, color='brown')
    ax.scatter(eary,earx, color='green')
    ax.scatter(mary,marx, color='red')
    ax.scatter(jupy,jupx, color='orange')
    ax.scatter(saty,satx, color='yellow')
    ax.scatter(uray,urax, color='cyan')
    ax.scatter(nepy,nepx, color='blue')
    
    
    #ax.axhline(y=0, color='k')
    #ax.axvline(x=0, color='k')
    ax.set_xlim([-45,45])
    ax.set_ylim([-45,45])
    
    print("Mercury:",round(merx,3),",",round(mery,3),
         "\nVenus:",round(venx,3),",",round(veny,3),
         "\nEarth:",round(earx,3),",",round(eary,3),
         "\nMars:",round(marx,3),",",round(mary,3),
         "\nJupiter:",round(jupx,3),",",round(jupy,3),
         "\nSaturn:",round(satx,3),",",round(saty,3),
         "\nUranus:",round(urax,3),",",round(uray,3),
         "\nNeptune:",round(nepx,3),",",round(nepy,3), 
         )
```

```python
def anasolarsys(vxo,vyo, xo,yo, t):
    #spacecraft
    x = np.zeros(t)
    y = np.zeros(t)
    vx = np.zeros(t)
    vy = np.zeros(t)
    ax = np.zeros(t)
    ay = np.zeros(t)
    a = np.zeros(t)
    
    #sun
    sun_ax = np.zeros(t) 
    sun_ay = np.zeros(t)
    sun_a = np.zeros(t)
    sun_r = np.zeros(t)
    sun_deltar = np.zeros(t)
    #mercury
    mer_ax = np.zeros(t) 
    mer_ay = np.zeros(t)
    mer_a = np.zeros(t)
    mer_r = np.zeros(t)
    mer_deltar = np.zeros(t)
    #venus
    ven_ax = np.zeros(t)
    ven_ay = np.zeros(t)
    ven_a = np.zeros(t)
    ven_r = np.zeros(t)
    ven_deltar = np.zeros(t)
    #earth
    ear_ax = np.zeros(t)
    ear_ay = np.zeros(t)
    ear_a = np.zeros(t)
    ear_r = np.zeros(t)
    ear_deltar = np.zeros(t)
    #mars
    mar_ax = np.zeros(t)
    mar_ay = np.zeros(t)
    mar_a = np.zeros(t)
    mar_r = np.zeros(t)
    mar_deltar = np.zeros(t)
    #jupiter
    jup_ax = np.zeros(t)
    jup_ay = np.zeros(t)
    jup_a = np.zeros(t)
    jup_r = np.zeros(t)
    jup_deltar = np.zeros(t)
    #saturn
    sat_ax = np.zeros(t)
    sat_ay = np.zeros(t)
    sat_a = np.zeros(t)
    sat_r = np.zeros(t)
    sat_deltar = np.zeros(t)
    #uranus
    ura_ax = np.zeros(t)
    ura_ay = np.zeros(t)
    ura_a = np.zeros(t)
    ura_r = np.zeros(t)
    ura_deltar = np.zeros(t)
    #neptune
    nep_ax = np.zeros(t)
    nep_ay = np.zeros(t)
    nep_a = np.zeros(t)
    nep_r = np.zeros(t)
    nep_deltar = np.zeros(t)
    
    h = 10**(-4)
    for i in range(t):
        if i == 0:
            #spacecraft position/velocity
            x[0] =  xo
            y[0] =  yo
            #velocities
            vx[0] = vxo
            vy[0] = vyo
            #accelerations
            ax[0] = 0
            ay[0] = 0
            a[0] = 0
        else:
            #sun acc
            sun_deltar[i] = ((x[i-1])**2+(y[i-1])**2)**(1/2)
            
            sun_a[i] = (-1*G*Ms)/((sun_deltar[i])**2)
            
            sun_ax[i] = (sun_a[i]*x[i-1])/((x[i-1]**2 + y[i-1]**2)**(1/2))                   
            sun_ay[i] = (sun_a[i]*y[i-1])/((x[i-1]**2 + y[i-1]**2)**(1/2))
            #mercury acc
            mer_deltar[i] = ((x[i-1]-mer.xo)**2+(y[i-1]-mer.yo)**2)**(1/2)
            
            mer_a[i] = (-1*G*mer.M)/((mer_deltar[i])**2)
            
            mer_ax[i] = (mer_a[i]*(x[i-1]-mer.xo))/(((x[i-1]-mer.xo)**2 + (y[i-1]-mer.yo)**2)**(1/2))                   
            mer_ay[i] = (mer_a[i]*(y[i-1]-mer.yo))/(((x[i-1]-mer.xo)**2 + (y[i-1]-mer.yo)**2)**(1/2))
            #venus acc
            ven_deltar[i] = ((x[i-1]-ven.xo)**2+(y[i-1]-ven.yo)**2)**(1/2)
            
            ven_a[i] = (-1*G*ven.M)/((ven_deltar[i])**2)
            
            ven_ax[i] = (ven_a[i]*(x[i-1]-ven.xo))/(((x[i-1]-ven.xo)**2 + (y[i-1]-ven.yo)**2)**(1/2))                   
            ven_ay[i] = (ven_a[i]*(y[i-1]-ven.yo))/(((x[i-1]-ven.xo)**2 + (y[i-1]-ven.yo)**2)**(1/2))
            #earth acc
            ear_deltar[i] = ((x[i-1]-ear.xo)**2+(y[i-1]-ear.yo)**2)**(1/2)
            
            ear_a[i] = (-1*G*ear.M)/((ear_deltar[i])**2)
            
            ear_ax[i] = (ear_a[i]*(x[i-1]-ear.xo))/(((x[i-1]-ear.xo)**2 + (y[i-1]-ear.yo)**2)**(1/2))                   
            ear_ay[i] = (ear_a[i]*(y[i-1]-ear.yo))/(((x[i-1]-ear.xo)**2 + (y[i-1]-ear.yo)**2)**(1/2))
            #mars acc
            mar_deltar[i] = ((x[i-1]-mar.xo)**2+(y[i-1]-mar.yo)**2)**(1/2)
            
            mar_a[i] = (-1*G*mar.M)/((mar_deltar[i])**2)
            
            mar_ax[i] = (mar_a[i]*(x[i-1]-mar.xo))/(((x[i-1]-mar.xo)**2 + (y[i-1]-mar.yo)**2)**(1/2))                   
            mar_ay[i] = (mar_a[i]*(y[i-1]-mar.yo))/(((x[i-1]-mar.xo)**2 + (y[i-1]-mar.yo)**2)**(1/2))
            #jupiter acc
            jup_deltar[i] = ((x[i-1]-jup.xo)**2+(y[i-1]-jup.yo)**2)**(1/2)
            
            jup_a[i] = (-1*G*jup.M)/((jup_deltar[i])**2)
            
            jup_ax[i] = (jup_a[i]*(x[i-1]-jup.xo))/(((x[i-1]-jup.xo)**2 + (y[i-1]-jup.yo)**2)**(1/2))                   
            jup_ay[i] = (jup_a[i]*(y[i-1]-jup.yo))/(((x[i-1]-jup.xo)**2 + (y[i-1]-jup.yo)**2)**(1/2))
            #saturn acc
            sat_deltar[i] = ((x[i-1]-sat.xo)**2+(y[i-1]-sat.yo)**2)**(1/2)
            
            sat_a[i] = (-1*G*sat.M)/((sat_deltar[i])**2)
            
            sat_ax[i] = (sat_a[i]*(x[i-1]-sat.xo))/(((x[i-1]-sat.xo)**2 + (y[i-1]-sat.yo)**2)**(1/2))                   
            sat_ay[i] = (sat_a[i]*(y[i-1]-sat.yo))/(((x[i-1]-sat.xo)**2 + (y[i-1]-sat.yo)**2)**(1/2))
            #uranus acc
            ura_deltar[i] = ((x[i-1]-ura.xo)**2+(y[i-1]-ura.yo)**2)**(1/2)
            
            ura_a[i] = (-1*G*ura.M)/((ura_deltar[i])**2)
            
            ura_ax[i] = (ura_a[i]*(x[i-1]-ura.xo))/(((x[i-1]-ura.xo)**2 + (y[i-1]-ura.yo)**2)**(1/2))                   
            ura_ay[i] = (ura_a[i]*(y[i-1]-ura.yo))/(((x[i-1]-ura.xo)**2 + (y[i-1]-ura.yo)**2)**(1/2))
            #neptune acc
            nep_deltar[i] = ((x[i-1]-nep.xo)**2+(y[i-1]-nep.yo)**2)**(1/2)
            
            nep_a[i] = (-1*G*nep.M)/((nep_deltar[i])**2)
            
            nep_ax[i] = (nep_a[i]*(x[i-1]-nep.xo))/(((x[i-1]-nep.xo)**2 + (y[i-1]-nep.yo)**2)**(1/2))                   
            nep_ay[i] = (nep_a[i]*(y[i-1]-nep.yo))/(((x[i-1]-nep.xo)**2 + (y[i-1]-nep.yo)**2)**(1/2))
            
            #spacecraft
            ax[i] = (sun_ax[i] + mer_ax[i] + ven_ax[i] + ear_ax[i] + mar_ax[i] + jup_ax[i] + sat_ax[i] + ura_ax[i] + nep_ax[i])
            ay[i] = (sun_ay[i] + mer_ay[i] + ven_ay[i] + ear_ay[i] + mar_ay[i] + jup_ay[i] + sat_ay[i] + ura_ay[i] + nep_ay[i])
            
            x[i] =  x[i-1] + vx[i-1]*h
            y[i] =  y[i-1] + vy[i-1]*h
        
            vx[i] =  vx[i-1] + ax[i]*h
            vy[i] =  vy[i-1] + ay[i]*h 
            
    fig, bx = plt.subplots(figsize=(30,30))
    bx.scatter(0,0, color='black'  )
    bx.scatter(mer.xo,mer.yo, color='grey'  )
    bx.scatter(ven.xo,ven.yo, color='brown' )
    bx.scatter(ear.xo,ear.yo, color='green')
    bx.scatter(mar.xo,mar.yo, color='red')
    bx.scatter(jup.xo,jup.yo, color='orange')
    bx.scatter(sat.xo,sat.yo, color='yellow')
    bx.scatter(ura.xo,ura.yo, color='cyan')
    bx.scatter(nep.xo,nep.yo, color='blue')
    bx.plot(x,y)
    
    #ax.axhline(y=0, color='k')
    #ax.axvline(x=0, color='k')
    bx.set_xlim([-45,45])
    bx.set_ylim([-45,45])
    print(bx)
#anasolarsys(vxo,vyo, xo,yo, t)
```
