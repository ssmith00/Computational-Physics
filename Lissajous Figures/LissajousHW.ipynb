```python
import math
import numpy as np
```

```python
def lissajous(fx,fy,Ax,Ay,phi,dt,N):
    X = list()
    Y = list()
    Z = list()
    
    ts = list()
    
    for i in range(N):
       
        t = i*dt
        
        ts.append(int(i*dt))
    
        X.append(Ax*math.cos(2*math.pi*fx*t))
        Y.append(Ay*math.cos(2*math.pi*fy*t + phi))
        Z.append(X[-1] + Y[-1])
        
    np.savetxt('lissajous', (X,Y,Z))
    return X, Y, Z, ts    
    
```

```python
lissajous(32,6,27,43,0.5,.1,50)
```

## b)

```python
def lissajous(fx,fy,Ax,Ay,phi,dt,N):
    
    X = np.array([])
    Y = np.array([])
    Z = np.array([])
    
    
    for i in range(N):
   
        t = i*dt
    
        np.append([Ax*math.cos(2*math.pi*fx*t)], X)
        np.append([Ay*math.cos(2*math.pi*fy*t + phi)], Y)
        np.append(X + Y, Z)
        
    np.save('lissajousnp', (X,Y,Z))
        
    
```

```python
X,Y,Z = lissajous(2,23,27,43,math.pi/2,.001,1000)
```
